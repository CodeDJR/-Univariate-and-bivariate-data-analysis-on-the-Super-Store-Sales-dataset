# Histograms

A histogram is a type of bar chart that represents the distribution of a dataset. It is used to visualize the frequency distribution of a continuous variable. The data is divided into bins (intervals), and the height of each bar represents the number of observations that fall within each bin.

**1.Distribution of sales**

```python
# Creating a histogram for the Sales column
plt.hist(df1['Sales'], bins=10, edgecolor='black')

# Adding titles and labels
plt.title('Histogram of Sales')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')

# Showing the plot
plt.show()
```

**Observations:**


1. The Histogram chart for the sales column shows that most sales are concentrated in the lower range, with a few outliers at the higher end.
2. The majority of sales fall between 0 and 500 dollars. There are a few sales that are much higher, but these are very rare.
3. So this suggest that the business may want to focus on increasing the number of smaller sales, as this is where the majority of their revenue comes from.



**2.Distribution of sales by Ship Mode.**

```python
# Creating histograms for each Ship Mode
ship_modes = df1['Ship Mode'].unique()
for mode in ship_modes:
    plt.hist(df1[df1['Ship Mode'] == mode]['Sales'], bins=10, edgecolor='black', alpha=0.5, label=mode)

# Adding titles and labels
plt.title('Histogram of Sales by Ship Mode')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.legend(title='Ship Mode')

# Showing the plot
plt.show()
```

**Observations:**


1.   The standard class ship mode has the highest sales.
2. Most of the customer preferred the standard ship mode due to purchase of more expensive items or large quantity, leading to higher overall sales.
3.   Also the standard ship mode has the outliers at the end as some customers have purchased very expensive items or large numbers of purchase from a single person. Also the first class ship mode has fewer outliers as only few customers have preferred it.





**3.Histogram of Sales by Category**

```python
df1.Category.value_counts()
```

2.   Furniture category has outliers as the sales is very less. The Business should increase their sales in furniture cateogry.

**Observations:**

1. The heatmap shows the correlation between Sales, Day, Month, and Year.

2. Sales has a weak positive correlation with Day and Year.

3. Sales has a moderate positive correlation with Month.

4. Day has a negligible correlation with Month and a moderate negative correlation with Year.

5. Month has a strong positive correlation with Year.
This suggests that as the day, month, and year increase, sales tend to increase as well.

6. However, the correlation is not very strong, so there may be other factors that are affecting sales.

7. The strong positive correlation between month and year is because they are related. As the year increases, the months also increase.

# Stacked Bar Chart

```python
# Creating a Stacked Bar Chart to compare Sales across different Region.

sales_by_region_category = df1.groupby(['Region', 'Category'])['Sales'].sum().reset_index()

# Create the stacked bar chart
fig, ax = plt.subplots(figsize=(10, 6))

regions = sales_by_region_category['Region'].unique()
categories = sales_by_region_category['Category'].unique()
bottom = np.zeros(len(regions))

for category in categories:
    sales = sales_by_region_category[sales_by_region_category['Category'] == category]['Sales']
    ax.bar(regions, sales, label=category, bottom=bottom)
    bottom += sales

# Set labels and title
ax.set_xlabel('Region')
ax.set_ylabel('Sales')
ax.set_title('Stacked Bar Chart of Sales by Region and Category')
ax.legend()

plt.show()
```

**Observation:**

1. The West region made higher sales and south region made the least and central nd east in the mid level.
2. The business should focus more on increasing the sales in the southern region.
